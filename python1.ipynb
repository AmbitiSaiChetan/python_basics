```python
import hashlib
email = input("Create email: ")
password = input("Create password: ")
stored_email = hashlib.sha256(email.encode()).hexdigest()
stored_pass = hashlib.sha256(password.encode()).hexdigest()
login_email = input("Enter email: ")
while hashlib.sha256(login_email.encode()).hexdigest() != stored_email:
    print("Wrong email! Try again.")
    login_email = input("Enter email: ")
login_pass = input("Enter password: ")
while hashlib.sha256(login_pass.encode()).hexdigest() != stored_pass:
    print("Wrong password! Try again.")
    login_pass = input("Enter password: ")
print("Login Successful!")
```
