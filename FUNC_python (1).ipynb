```python
l=[1,2,3,4,5]
def square(x):
    return x**2
def cube(x):
    return x**3
def transform(f,l):
    output=[]
    for i in l:
        output.append(f(i))
    return output
```

## The reduce(fun,seq) function is used to apply a particular function passed in

```python
import functools
from functools import reduce
l=[1,2,3,4,5]
functools.reduce(lambda x,y:x+y,l)
```
